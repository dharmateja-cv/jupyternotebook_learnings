#### Day-to-day percentage change of the stock

Daily percentage change in the price of the stock is calculated on the basis of percentage change between 2 consecutive days closing prices. Let’s say if the closing price of the stock yesterday was 600 and today the stock closed as 500. So, the percentage change is 10%. i.e. ((600–500) / 600)*100. 

Accordingly, we’ll introduce a new column "Day_perc_chg" denoting the daily returns in the price of the stock. 




```python
sbi_df_perc_chg = (sbi_df['close ']- sbi_df['PREV. CLOSE ']  )/sbi_df['PREV. CLOSE ']

sbi_df['Day_Perc_Chg']  = sbi_df_perc_chg *100

sbi_df.head()
```

```python
#Drop the Null value column from the Data frame if any
sbi_df.dropna(inplace = True, axis = 0)
```

```python
sbi_df
```

```python
sbi_df['Day_Perc_Chg'].plot(figsize = (12, 6), fontsize = 12);
```

We can observe that for most of the days, the returns are between -5% to 5% with few spikes in between crossing 12% mark on both the sides.

### We will Plot daily returns distribution histogram

```python
sbi_df['Day_Perc_Chg']
```

## Q3  Can we find the Trend of the SBI stock for the entire 9 months period? 

# Trend Analysis

#### Next we add a new column ‘Trend’ whose values are based on the day-to-day percentage change we calculated above. 
##### Trend is determined from below relationship —

##### *Note : The values  considred for trend segregation are purely to differentiate the stock performance and this is not the approved/mandatory ranges to calculate the Trend 

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return 'Slight or No change'
    elif x > 0.5 and x <= 1:
        return 'Slight Positive'
    elif x > -1 and x <= -0.5:
        return 'Slight Negative'
    elif x > 1 and x <= 3:
        return 'Positive'
    elif x > -3 and x <= -1:
        return 'Negative'
    elif x > 3 and x <= 7:
        return 'Among top gainers'
    elif x > -7 and x <= -3:
        return 'Among top losers'
    elif x > 7:
        return 'Bull run'
    elif x <= -7:
        return 'Bear drop'
```

```python
sbi_df['Trend']= np.zeros(sbi_df['Day_Perc_Chg'].count())
```

```python
np.zeros(sbi_df['Day_Perc_Chg'].count())
```

```python
sbi_df['Trend']= sbi_df['Day_Perc_Chg'].apply(lambda x:trend(x))
```

```python
sbi_df.head()
```

```python
sbi_pie_data = sbi_df.groupby('Trend')
```

```python
pie_label = sorted([i for i in sbi_df.loc[:, 'Trend'].unique()])
```

```python
plt.pie(sbi_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 1)
plt.show()
```

### From the above pie chart we can see that 

SBI stock has
* 22.1 % of the times the stock was with Postive Trend
* 9.9  % of the times the stock was Among top Gainers
